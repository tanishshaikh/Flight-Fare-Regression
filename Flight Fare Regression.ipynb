```python
lis=list(data['Duration'])
```

```python
new_lis=[]
for i in lis:
    if (len(i.split(' '))==1):
        if 'h' in i:
            i=i+' 0m'
        elif 'm' in i:
            i='0h '+i

    new_lis.append(i)
```
